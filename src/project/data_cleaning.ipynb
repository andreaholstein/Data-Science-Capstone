```python
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
import scipy
import PIL
import requests
```

```python
# load data, removing spaces in column headers
columns = ["Customer_ID", "Age", "Gender", "Item_Purchased", "Item_Category", "Purchase_Amount_(USD)", "Location", "Color", "Season", "Review_Rating", "Subscription_Status", "Shipping_Type", "Discount_Applied", "Promo_Code_Used", "Previous_Purchases", "Payment_Method", "Frequency_of_Purchases"]
df = pd.read_csv("../data/Customers3.csv", header = 0, names = columns)
df
```

```python
# typo checking!
for i in df.columns:
    print(df[i].unique()) # ex: print(df["Gender"].unique())
```

```python
# data cleaning - checking for null values
iN = df.isnull()
print(iN)
```

```python
# checking for missing values, by column indexes and iterating over its rows
all_cols = df.columns

for i in all_cols:
    null_rows = df[df[i].isnull()].index.tolist()
    empty_rows = df[df[i] == ''].index.tolist()
    print(null_rows, empty_rows)
```

```python
# dropping empty rows and checking if dataframe has changed shape - it hasn't = no empties!
# df.shape # (3900, 17)
drop_missing = df.dropna()
drop_missing.shape # (3900, 17)
```
