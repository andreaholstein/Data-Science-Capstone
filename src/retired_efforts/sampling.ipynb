```python
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd 
import scipy
import PIL
import requests
```

```python
# load data, removing spaces in column headers
columns = ["Customer_ID", "Age", "Gender", "Item_Purchased", "Item_Category", "Purchase_Amount_(USD)", "Location", "Color", "Season", "Review_Rating", "Subscription_Status", "Shipping_Type", "Discount_Applied", "Promo_Code_Used", "Previous_Purchases", "Payment_Method", "Frequency_of_Purchases"]
df = pd.read_csv("../data/Customers3.csv", header = 0, names = columns)
# df
```

Identifying Shopper Loyalty by Sorting the Original Dataframe by Previous Purchase Volume and Purchase Amount ($)

```python
# sort original dataframe by volume of previous purchases and purchase amounts (highest first) to identify most loyal shoppers
df_sorted = df.sort_values(by=['Previous_Purchases', 'Purchase_Amount_(USD)'], ascending = False)
# df_sorted = df.sort_values(by=['Previous_Purchases', 'Purchase_Amount_(USD)', 'Frequency_of_Purchases'], ascending = False) # can add in frequency of purchases
df_sorted
```

```python
# create smaller sample of 500 top values from the sorted dataframe
top_500 = df_sorted[:500]
top_500
```
